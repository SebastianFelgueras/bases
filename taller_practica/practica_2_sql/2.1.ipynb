### Ejercicio 1

#### a)

```sql
SELECT FirstName, LastName FROM Customer
WHERE Country = 'Brazil';
```

#### b)

```sql
SELECT c.CustomerId, i.InvoiceDate, i.InvoiceId FROM Customer c
INNER JOIN Invoice i ON i.CustomerId = c.CustomerId;
```

#### c)

```sql
SELECT t.Name, a.Name FROM Track t
LEFT JOIN Album al ON t.AlbumId = al.AlbumId 
LEFT JOIN Artist a ON al.ArtistId = a.ArtistId;
```

#### e)

```sql
SELECT p.Name FROM Playlist p
INNER JOIN PlaylistTrack pt ON p.PlaylistId = pt.PlaylistId
INNER JOIN Track t ON t.TrackId = pt.TrackId
INNER JOIN Album a ON t.AlbumId = a.AlbumId
INNER JOIN Artist ar ON a.ArtistId = ar.ArtistId
WHERE ar.Name = 'Iron Maiden'
GROUP BY p.PlaylistId, p.Name
HAVING COUNT(t.TrackId) > 10;
```

#### f)

```sql
SELECT p.Name, COUNT(DISTINCT t.AlbumId) n_albumes FROM Playlist p
INNER JOIN PlaylistTrack pt ON p.PlaylistId = pt.PlaylistId
INNER JOIN Track t ON t.TrackId = pt.TrackId
GROUP BY p.PlaylistId, p.Name;
```

#### g)

```sql
SELECT DISTINCT e.FirstName, e.LastName FROM Employee e
INNER JOIN Customer c ON c.SupportRepId = e.EmployeeId
INNER JOIN Invoice i ON c.CustomerId = i.CustomerId
INNER JOIN InvoiceLine il ON il.InvoiceId = i.InvoiceId 
WHERE e.BirthDate < '2000-05-22'
GROUP BY e.EmployeeId, e.FirstName, e.LastName, i.InvoiceId
HAVING COUNT(il.InvoiceLineId) > 10;
```

#### h)

```sql
SELECT c.CustomerId, i.InvoiceDate, i.InvoiceId FROM Customer c
LEFT JOIN Invoice i ON i.CustomerId = c.CustomerId;
```

#### l)

```sql
WITH compras_por_factura AS (SELECT c.CustomerId, SUM(il.Quantity) Cantidad FROM Customer c
LEFT JOIN Invoice i ON c.CustomerId = i.CustomerId
LEFT JOIN InvoiceLine il ON i.InvoiceId = il.InvoiceId
GROUP BY c.CustomerId, i.InvoiceId)

SELECT cf.CustomerId, AVG(cf.Cantidad) Promedio  FROM compras_por_factura cf
GROUP BY cf.CustomerId;
```

#### m)

```sql
WITH cantidad AS (SELECT e.EmployeeId, COUNT(t.TrackId) Cantidad FROM Employee e
LEFT JOIN Customer c ON c.SupportRepId = e.EmployeeId
LEFT JOIN Invoice i ON i.CustomerId = c.CustomerId
LEFT JOIN InvoiceLine il ON il.InvoiceId = i.InvoiceId
LEFT JOIN Track t ON il.TrackId = t.TrackId
LEFT JOIN Genre g ON g.GenreId = t.GenreId
WHERE g.Name = 'Rock'
GROUP BY e.EmployeeId)

SELECT e.EmployeeId, COALESCE(c.Cantidad,0)  FROM Employee e
LEFT JOIN cantidad c ON c.EmployeeId = e.EmployeeId
ORDER BY e.EmployeeId
;
```
